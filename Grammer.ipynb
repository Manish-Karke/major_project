```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT model and tokenizer
model_name = 'gpt2'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

def check_grammar(sentence, num_return_sequences=5):
    # Tokenize input sentence
    inputs = tokenizer.encode(sentence, return_tensors='pt')
    # Generate suggestions with sampling
    outputs = model.generate(
        inputs,
        max_length=50,
        num_return_sequences=num_return_sequences,
        do_sample=True,  # Enable sampling
        top_k=50,  # Consider the top 50 tokens by probability
        top_p=0.95  # Consider tokens with a cumulative probability of 0.95
    )
    suggestions = [tokenizer.decode(output, skip_special_tokens=True) for output in outputs]
    return suggestions

# Example sentence with grammar errors
sentence = "She don't like apples because it taste weird."
suggestions = check_grammar(sentence)

print("Original Sentence:", sentence)
print("Grammar Suggestions:")
for i, suggestion in enumerate(suggestions):
    print(f"{i+1}. {suggestion}")
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

model = T5ForConditionalGeneration.from_pretrained('t5-base')
tokenizer = T5Tokenizer.from_pretrained('t5-base')

text = "I love eat apples daily."
input_ids = tokenizer.encode(text, return_tensors='pt')
outputs = model.generate(input_ids)

corrected_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(corrected_text)
```
